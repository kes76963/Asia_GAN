```python
# 제목 추출
def tbt(category_num, page) :
    
    url = f'https://www.10x10.co.kr/shopping/category_main.asp?rect=&prvtxt=&rstxt=&extxt=&sflag=n&disp={category_num}&cpg={page}&chkr=False&chke=False&mkr=&sscp=Y&psz=40&srm=be&iccd=0&styleCd=&attribCd=&icoSize=S&arrCate=&deliType=&minPrc=&maxPrc=&lstDiv=list&ab='
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36'}
    html = requests.get(url, headers=headers)
    title = BeautifulSoup(html.text, 'html.parser')
    # keyword = soup_rank.select('div > div:nth-of-type(2) > div:nth-of-type(4) > div.ibox.float-e-margins > div ')
    keywords = title.select('p.pdtName.tPad07')
    
```

```python
key_list = []

for i in range(1,151) :
    tbt(119,i)
    #6개는 행사 상품
    for k in keywords[6:78]:
        keyword = k.text
        key_list.append(keyword)
    if i % 10 == 0 : 
        print(i,'페이지 완료')
        
        
print('finish')
```

```python
#[] () 안에 있는 문자열 삭제
import re
import pandas as pd

key_list2 = []
df_section_title = pd.DataFrame()

for key in key_list :
    if '[' and '(' in key :
        key = re.sub(r'\[[^)]*\]', '', key)
        key_re = re.sub(r'\([^)]*\)', '', key)
        key_list2.append(key_re)        
    elif '[' in key :
        key_re = re.sub(r'\[[^)]*\]', '', key)
        key_list2.append(key_re)
    elif '(' in key :
        key_re = re.sub(r'\([^)]*\)', '', key)
        key_list2.append(key_re)
    else :
        key_list2.append(key_re)

print(key_list2)
print('(),[]삭제 완료')
df_section_title = pd.DataFrame(key_list2, columns=['title'])
df_section_title['category'] = 'Food'
df_section_title.to_csv('project_4/food.csv')

print('csv 저장완료')

```
